## C

C é uma linguagem de programação de baixo nível, criada na década de 1970 por Dennis Ritchie nos laboratórios da Bell, com o objetivo de desenvolver sistemas operacionais, como o Unix. Ela é conhecida por ser eficiente, pois oferece controle direto sobre a memória e o hardware do computador, o que a torna ideal para o desenvolvimento de sistemas, drivers e outros softwares que precisam de desempenho elevado.

Algumas características importantes da linguagem C são:

- **Portabilidade**: Programas escritos em C podem ser executados em diferentes plataformas com pouca modificação.
- **Controle de memória**: A linguagem oferece acesso direto ao gerenciamento de memória, o que permite um controle detalhado dos recursos do sistema.
- **Estruturada**: C permite que os programas sejam organizados em funções, facilitando a legibilidade e manutenção do código.
- **Extensa biblioteca padrão**: A linguagem inclui uma vasta biblioteca com funções para manipulação de strings, arquivos, e outros elementos essenciais.

C é frequentemente vista como a "base" de outras linguagens modernas, como C++, C# e Java, devido à sua influência no design de linguagens posteriores.


# hello world

Em C, para imprimir a mensagem hello world, é necessário utilizar a função printf, que é uma função da biblioteca padrão stdio.h. 

include <stdio.h> é uma diretiva de pré-processamento que inclui o conteúdo do arquivo de cabeçalho stdio.h no programa. stdio.h é um arquivo de cabeçalho padrão em C. stdio significa standard input/output

A função printf é utilizada para imprimir mensagens na tela. O f significa format, que indica que a função printf aceita argumentos formatados. O comando \n é utilizado para pular uma linha pois a função print, em C não possui esse comportamento por padrão

As chaves {} são utilizadas para delimitar o corpo da função main. O corpo da função main é o bloco de código que será executado quando o programa for executado. Usamos ponto e vírgula ; para finalizar uma instrução em C.

```c
#include <stdio.h>

int main(void)
{
    printf("hello, world\n");
}
```

1. **Biblioteca Padrão**:
   ```c
   #include <stdio.h>
   ```
   O `#include` importa bibliotecas da linguagem, neste caso, a `<stdio.h>`, que é uma biblioteca padrão usada para operações de entrada e saída, como a função `printf`. Isso demonstra o uso das **bibliotecas padrão** do C, que fornecem funções essenciais.

2. **Função `main`**:
   ```c
   int main(void)
   ```
   Toda execução de um programa em C começa pela função `main`. O tipo de retorno dessa função é `int`, o que significa que ela retorna um valor inteiro ao sistema operacional quando o programa termina. Isso reflete a natureza **estruturada** da linguagem, onde os programas são organizados em funções.

3. **Função `printf`**:
   ```c
   printf("hello, world\n");
   ```
   A função `printf` é usada para imprimir uma string na tela. Aqui, a string `"hello, world\n"` é impressa, e o `\n` adiciona uma nova linha no final. Este exemplo mostra o uso de **funções da biblioteca padrão** para interagir com o usuário.

4. **Controle Explícito de Tipos**:
   ```c
   int main(void)
   ```
   Em C, é necessário especificar o tipo de dados das variáveis e funções. Aqui, o `int` indica que a função `main` retornará um valor inteiro. Isso reflete a característica de **tipagem estática**, em que os tipos de dados devem ser declarados explicitamente.

# Compiladores


Uma vez que salvamos o código que escrevemos, que é chamado de código-fonte, precisamos convertê-lo em código de máquina, instruções binárias que o computador entende diretamente. Usamos um programa chamado compilador para compilar nosso código-fonte em código de máquina. O compilador verifica o código-fonte para garantir que ele esteja escrito corretamente e, em seguida, gera um arquivo executável que você pode executar. C é uma linguagem de programação compilada, o que significa que você precisa de um compilador para transformar seu código-fonte em código de máquina.

Para compilar um programa C, em nosso terminal digitamos o comando `make` seguido do nome do arquivo que contém o código-fonte. Por exemplo, se o arquivo que contém o código-fonte for chamado `hello.c`, digitamos `make hello`. O comando `make` chama o compilador C, que é chamado de **gcc** (GNU Compiler Collection), para compilar o código-fonte em um arquivo executável chamado `hello`. O arquivo executável é chamado de `hello` porque tem o mesmo nome que o arquivo de código-fonte, mas sem a extensão `.c`.

Após compilar o código-fonte, podemos executar o arquivo executável digitando `./` seguido do nome do arquivo executável. Por exemplo, se o arquivo executável for chamado `hello`, digitamos `./hello` para executá-lo. Usamos `./` para indicar que o arquivo executável está no diretório atual.

# String

Strings são cadeias de caracteres, podendo conter zero ou mais caracteres, sendo eles letras, números, símbolos, etc.

Em C, strings são implementadas como arrays de caracteres, ou seja, cada caractere da string é armazenado em uma posição de memória diferente. Uma string em C sempre termina com um caractere especial chamado null terminator ('\0'), que indica o fim da string. Isso é essencial para que o compilador saiba onde a string termina na memória.

Representação de Strings em C:
As strings são representadas entre aspas duplas ("string"). Quando uma string é declarada, o compilador automaticamente adiciona o caractere '\0' no final para indicar o fim da string.
