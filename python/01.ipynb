```python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
indicadores = ['CUPONS', 'TKM', 'PM', 'PA']
variacoes = [-6.1, -0.5, 9.0, -8.7]  # em porcentagem
cores = ['#e74c3c', '#e74c3c', '#2ecc71', '#e74c3c']  # vermelho para negativo, verde para positivo

# Criando o gráfico
fig, ax = plt.subplots(figsize=(10, 6))

# Criando barras horizontais
bars = ax.barh(indicadores, variacoes, color=cores, height=0.6)

# Adicionando os valores nas barras
for bar in bars:
    width = bar.get_width()
    label_x = width + 0.5 if width >=0 else width - 0.8
    ax.text(label_x, bar.get_y() + bar.get_height()/2, 
            f'{width:.1f}%', 
            va='center', ha='center' if width <0 else 'left',
            color='black', fontweight='bold')

# Linha vertical no zero
ax.axvline(x=0, color='black', linestyle='-', linewidth=0.8)

# Configurações do gráfico
ax.set_xlim(-10, 10)
ax.set_xticks(np.arange(-10, 11, 2))
ax.set_xlabel('Variação Percentual (%)')
ax.set_title('Variação dos Indicadores (2025 vs 2024)', pad=20, fontweight='bold')
ax.grid(axis='x', linestyle='--', alpha=0.6)

# Removendo bordas
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.tight_layout()
plt.show()
```
