# 📂 Manipulação de Arquivos de Texto

Um dos recursos mais importantes na programação é a capacidade de **ler e salvar dados** em arquivos no seu computador. Imagine que você conectou em uma API, puxou dados importantes, mas quando o programa fecha, tudo some!

Para resolver isso, usamos blocos de manipulação de arquivos no Python, principalmente a função `open()` e o gerenciador de contexto `with`.


### 1. Criando e Escrevendo em um Arquivo (`w` - Write)

Usamos o modo de abertura `'w'` (Write) para criar um arquivo ou sobrescrever o que já existe.


```python
with open('meu_diario.txt', 'w', encoding='utf-8') as arquivo:
    arquivo.write('Primeira linha do meu arquivo!\n')
    arquivo.write('Programar em Python é muito legal.\n')

print('Arquivo criado com sucesso!')
```

### 2. Lendo um Arquivo (`r` - Read)

Para ler, usamos o modo `'r'` (Read), que inclusive é o padrão se você não digitar nada.


```python
with open('meu_diario.txt', 'r', encoding='utf-8') as arquivo:
    conteudo = arquivo.read()

print('--- Lendo o arquivo todo ---')
print(conteudo)
```

Se o arquivo for muito grande (milhares de linhas), ler tudo de uma vez pode travar a memória. Nesses casos, iteramos linha por linha no loop `for`:


```python
with open('meu_diario.txt', 'r', encoding='utf-8') as arquivo:
    for linha in arquivo:
        # end='' evita pular linha duas vezes, pois a linha já vem com a quebra no final
        print('--->', linha, end='')
```

### 3. Adicionando texto no final (`a` - Append)

Se abrirmos de novo com `'w'`, o Python **apaga** o arquivo antigo e cria do zero. Para **adicionar** informações no final do arquivo sem apagar o passado, usamos `'a'` (Append).


```python
with open('meu_diario.txt', 'a', encoding='utf-8') as arquivo:
    arquivo.write('Esta é uma linha adicionada depois!\n')

print('Nova linha adicionada!')
```

```python
with open('meu_diario.txt', 'r', encoding='utf-8') as arquivo:
    print(arquivo.read())
```

### 4. Lendo e Escrevendo Listas (`writelines` e `readlines`)

Constantemente você vai ter uma lista de informações (ex: produtos, nomes) e precisa jogá-los todos de uma vez no arquivo.


```python
compras = ['Maçã\n', 'Banana\n', 'Cenoura\n', 'Pão\n']

with open('lista_de_compras.txt', 'w', encoding='utf-8') as arquivo:
    arquivo.writelines(compras)
```

```python
with open('lista_de_compras.txt', 'r', encoding='utf-8') as arquivo:
    linhas_lidas = arquivo.readlines()
    
print(linhas_lidas)
```

Repare que no retorno acima os itens vieram com o termo `\n` (quebra de linha) no final de cada palavra. Se quiser limpar isso de uma lista, use a string `.strip()`:


```python
produtos_limpos = []
for produto in linhas_lidas:
    produtos_limpos.append(produto.strip())
    
print(produtos_limpos)
```

### 5. Boas Práticas: O poder do `with`

Sempre use o `with open(...) as ...`. Esse bloco é chamado de **Gerenciador de Contexto** (Context Manager).

Antigamente, você precisaria abrir o arquivo assim `f = open('ok.txt')` e, ao final de tudo, **lembrar** de usar `f.close()`. Se desse um erro no programa no meio do caminho e você esquecesse o `.close()`, o arquivo poderia ficar preso na memória e corromper.

O `with` fecha o arquivo **automaticamente e em segurança** assim que a indentação acaba, mesmo que o código apresente algum erro lá dentro!
