## Introdução a programação
---
Programar é fundamentalmente *``comunicação``*. Precisamos nos tornar bons comunicadores para expressar nossas ideias de forma clara e eficaz. Ao criar um código, estamos instruindo o computador a realizar uma tarefa por meio de um *algoritimo*, por sua vez é *um conjunto de instruções que são executadas em uma ordem específica para atingir um objetivo*. O código é a representação escrita desse algoritimo. Para que o computador entenda o que queremos que ele faça, precisamos escrever o código de forma clara e precisa, sem ambiguidade. A clareza do código é fundamental para que possamos entender o que ele faz e para que outras pessoas possam ler e compreender o que foi feito. A eficácia do código está relacionada à sua capacidade de realizar a tarefa proposta de forma correta e eficiente. Para escrever um código eficaz, precisamos considerar a sua ``legibilidade``, a sua ``eficiência`` e a sua ``manutenibilidade``.

- A legibilidade está relacionada à facilidade de leitura e compreensão do código. Um código legível é aquele que pode ser facilmente compreendido por outras pessoas, mesmo que elas não tenham conhecimento prévio sobre o assunto.

- A eficiência está relacionada à capacidade do código de realizar a tarefa proposta de forma rápida e sem desperdício de recursos. Um código eficiente é aquele que utiliza os recursos disponíveis de forma otimizada, evitando retrabalho e desperdício de tempo e memória. 

- A manutenibilidade está relacionada à facilidade de modificar e atualizar o código. Um código manutenível é aquele que pode ser facilmente alterado para corrigir erros, adicionar novas funcionalidades ou adaptar-se a novos requisitos.

Quando falamos de programação, é essencial compreender diversas camadas para dominar uma linguagem específica.
Primeiramente, entender a organização estrutural é fundamental. Cada linguagem de programação é como um idioma para computadores, com sua própria sintaxe e conjunto de normas.
 Assim como aprendemos a nos comunicar em um idioma natural, aprender a programar envolve conhecer a estrutura, o vocabulário, as regras e o significado das instruções.

Seja uma linguagem natural, como o português, ou uma linguagem de programação, todas compartilham elementos fundamentais. 


1. **Alfabeto**:  
   Toda linguagem, humana ou de máquina, é construída com base em um conjunto de símbolos. Na linguagem natural, como o português, temos letras, números e sinais de pontuação. Já nas linguagens de programação, o alfabeto inclui letras, números e caracteres especiais como `+`, `-`, `{`, `}`, `;`. Em linguagens de máquina, como o código binário, o alfabeto é reduzido a dois símbolos: `0` e `1`. Computadores utilizam a linguagem binária porque ela corresponde aos dois estados básicos dos circuitos eletrônicos: ligado (1) e desligado (0). Por exemplo, ao pressionar uma tecla no computador, essa ação é convertida em uma sequência de zeros e uns que o processador entende.

   *Em Python, a codificação padrão utilizada para representar caracteres é o ``UTF-8``, um formato amplamente usado que suporta uma grande variedade de caracteres, incluindo os do conjunto ``ASCII (American Standard Code for Information Interchange)``. O ASCII é uma codificação mais antiga, que usa 7 bits para representar até 128 caracteres, incluindo letras em inglês, números, e alguns símbolos. Já o UTF-8 é capaz de codificar mais de 1 milhão de caracteres, abrangendo praticamente todos os símbolos usados globalmente, como acentos, caracteres especiais e ideogramas de várias línguas.*

   *Isso significa que, em Python, você pode usar tanto os caracteres mais simples do ASCII quanto caracteres mais complexos e internacionais que fazem parte do UTF-8. Isso é especialmente útil ao programar em contextos onde diferentes idiomas ou símbolos são necessários. Por exemplo, você pode usar caracteres como "ç", "á", "ñ", e até emojis 😊 diretamente no seu código, graças à compatibilidade com UTF-8.*


2. **Vocabulário (Dicionário)**:  
   Toda linguagem tem seu próprio conjunto de palavras e expressões válidas. Em português, por exemplo, temos palavras como "casa", "carro" e "comida". Nas linguagens de programação, o vocabulário é composto por palavras-chave e funções predefinidas. Em Python, as palavras-chave são termos reservados que possuem significados específicos e não podem ser usados como nomes de variáveis ou funções. Exemplos de palavras-chave em Python incluem `if`, `else`, `for`, `while`, `def`, `return`, `True`, e `False`.

   Comparado a uma linguagem humana como o português, que possui centenas de milhares de palavras, o dicionário de Python é muito menor. Atualmente, Python tem cerca de 35 palavras-chave. Esse vocabulário compacto facilita o aprendizado, pois o programador precisa memorizar um conjunto limitado de termos para começar a programar. 

   **Palavras-chave** são fundamentais para a estrutura do código em Python, pois são usadas para definir condições, laços, funções, e muito mais. Elas são as “palavras de comando” que orientam a lógica do programa. Por exemplo:

   - `if` e `else` são usados para criar condições.
   - `for` e `while` são usados para criar loops.
   - `def` define funções, e `return` retorna valores dessas funções.

3. **Sintaxe**:  
   As regras que definem como as palavras podem ser combinadas formam a sintaxe da linguagem. Em português, há regras gramaticais que determinam a ordem das palavras em uma frase, como "O gato está no telhado". Em programação, a sintaxe define como os comandos devem ser escritos para serem compreendidos pela máquina. Em Python, por exemplo, a indentação é essencial para definir blocos de código. Quando a sintaxe é violada, como ao esquecer uma indentação ou usar símbolos incorretos, o programa gera erros que impedem sua execução. Por exemplo, esquecer de adicionar dois pontos após uma declaração de condição (if) ou usar parênteses incorretamente em uma função são erros de sintaxe típicos. Esses erros causam interrupções imediatas, alertando o programador de que algo está errado. Para que o código seja executado corretamente, essas regras sintáticas devem ser seguidas à risca, assim como respeitar a gramática em uma língua natural garante que a comunicação seja clara e compreensível.

4. **Semântica**:  
   Mesmo que uma frase esteja correta gramaticalmente, ela pode não fazer sentido se o significado não estiver claro. Por exemplo, "O peixe dirigiu o carro" é sintaticamente correta, mas semanticamente ilógica. Em programação, a semântica diz respeito ao comportamento do código. Um código pode ser sintaticamente correto, mas se sua lógica for errada, ele produzirá resultados indesejados.

### Exemplos

Entender como esses conceitos se aplicam na prática ajuda a solidificar o conhecimento. Vamos ver alguns exemplos simples:

- Imagine o momento em que você liga uma luz. Esse ato pode ser representado como "1" (ligado) e "0" (desligado). Em um computador, todas as operações, desde o movimento do cursor até a reprodução de um vídeo, são uma sequência complexa de operações binárias. Essas operações são baseadas em bits, a menor unidade de informação em um computador, que pode ser 0 ou 1. Vários bits agrupados formam um byte, que pode representar diferentes tipos de dados, como uma letra, um número ou uma cor.

Um exemplo interessante é o funcionamento das telas de TVs ou monitores. Cada pixel na tela é composto por três pequenos LEDs: um vermelho, um verde e um azul (RGB). A combinação desses LEDs, variando entre diferentes intensidades de "ligado" (1) e "desligado" (0), cria uma vasta gama de cores que forma a imagem que você vê. Quando todos os LEDs de um pixel estão desligados (0, 0, 0), a cor preta é mostrada. Quando todos estão no nível máximo (255, 255, 255), a cor branca aparece. A mistura dessas intensidades gera todas as outras cores. Por exemplo, ao ligar apenas o LED vermelho e o verde em alta intensidade, você obtém a cor amarela. Este processo, invisível ao olho humano, é um exemplo prático de como a linguagem binária, simples na base, cria resultados visuais complexos e ricos.

- Imagine uma máquina de lavar inteligente que ajusta o tempo de lavagem de acordo com o peso da roupa. Quando a lavagem começa, o software da máquina faz uma série de verificações automáticas. Por exemplo, ele verifica se "A roupa está pesada?". Se a resposta for sim, o software aumenta o tempo de lavagem; se não, ele segue o tempo padrão. Essas decisões são baseadas em instruções lógicas, como condições do tipo if e else, programadas no software. De forma invisível para o usuário, o sistema analisa várias condições e ajusta o funcionamento da máquina automaticamente, garantindo que cada lavagem seja feita de maneira otimizada, sem a necessidade de intervenção humana.

Dominar a programação envolve muito mais do que aprender uma lista de comandos. Assim como aprender um novo idioma, é preciso entender como as partes se relacionam para formar um todo coerente. Ao entender o alfabeto, o vocabulário, a sintaxe e a semântica de uma linguagem, você estará melhor preparado para criar soluções robustas e funcionais que resolvem problemas no mundo real.

A comunicação, seja entre pessoas ou entre humanos e máquinas, depende desses princípios fundamentais. Portanto, tanto a linguagem natural quanto a de programação compartilham a mesma base de organização, o que torna o aprendizado de programação uma habilidade acessível e poderosa para transformar ideias em realidade.

## Apresentando o Python
---

A linguagem de máquina é a forma mais básica de comunicação com o processador de um computador, sendo diretamente compreendida e executada por ele. Ela é composta por instruções binárias, ou seja, uma sequência de 0s e 1s, que correspondem a operações simples como adição, subtração, ou a movimentação de dados de um local para outro na memória. No entanto, essa linguagem é de baixo nível, o que significa que é extremamente difícil para os humanos entenderem e escreverem diretamente.

Para facilitar esse processo, foram desenvolvidas as linguagens de programação de alto nível. Essas linguagens são mais próximas da forma como os humanos pensam e se expressam, tornando a escrita de programas muito mais acessível e compreensível para os programadores. Ao usar uma linguagem de alto nível, o desenvolvedor pode escrever comandos de maneira intuitiva, sem se preocupar diretamente com os detalhes binários.

Entretanto, o computador não compreende diretamente essas linguagens de alto nível. Antes que um programa escrito dessa forma possa ser executado, ele precisa ser convertido para a linguagem de máquina. Esse processo é realizado por compiladores ou interpretadores, que traduzem o código de alto nível para a sequência binária que o processador entende, permitindo assim a execução das instruções no hardware.

**``Python``**: É uma linguagem de programação de alto nível que ganhou popularidade devido à sua sintaxe clara e legibilidade. Muitas pessoas acham que a estrutura do código Python se assemelha à linguagem inglesa, tornando-a relativamente fácil de aprender e entender para iniciantes. No entanto, é importante notar que, apesar de sua semelhança com a linguagem humana, Python ainda é uma linguagem de programação com regras e sintaxe específicas que devem ser seguidas para escrever código válido.
Python é uma linguagem de programação popular criada por Guido van Rossum em 1991. É amplamente utilizado em diversas áreas, incluindo desenvolvimento web (server-side), desenvolvimento de software, matemática e script do sistema.

Python pode ser usado para criar aplicações web em servidores, desenvolver fluxos de trabalho em conjunto com outros softwares, conectar-se a sistemas de banco de dados, manipular arquivos, lidar com big data, realizar matemática complexa, prototipagem rápida e desenvolvimento de software pronto para produção.

Algumas razões para escolher Python incluem sua capacidade de funcionar em diferentes plataformas, sua sintaxe simples e legível, sua capacidade de escrever programas com menos linhas de código em comparação com outras linguagens, sua execução rápida devido ao sistema interpretador e sua capacidade de ser tratado de forma processual, orientada a objetos ou funcional.

A versão principal mais recente é o Python 3, embora o Python 2 ainda seja popular. Python pode ser escrito em editores de texto ou em Ambientes de Desenvolvimento Integrado (IDEs) como PyCharm e VS Code.

A sintaxe de Python é projetada para legibilidade e usa novas linhas para completar um comando. Ela também depende de recuo (espaço em branco) para definir o escopo, como o escopo de loops, funções e classes, ao contrário de outras linguagens que geralmente usam colchetes para essa finalidade.

O Python pode já estar instalado em muitos PCs e Macs. Para verificar se está instalado, siga estas instruções:

- No Windows, procure por Python na barra de busca ou execute `python --version` no prompt de comando (cmd.exe).
- No Linux ou Mac, abra o Terminal e digite `python --version`.

Se não estiver instalado, você pode baixá-lo gratuitamente no site oficial python.org.

Para começar a programar em Python, siga este guia rápido:

1. Escreva seu código Python em um arquivo de texto (com extensão .py).
2. Execute o arquivo Python no interpretador Python.

Por exemplo, vamos criar um arquivo chamado "helloworld.py" com o seguinte conteúdo:

```python

```


### Introdução ao Python

A estrutura de um código Python é fundamental para garantir que o programa seja executado corretamente e de forma compreensível tanto para o desenvolvedor quanto para o interpretador da linguagem.

Aqui estão alguns aspectos importantes da estrutura de um código Python:

1. **Indentação**: Em Python, a indentação é crucial para definir a estrutura do código. Blocos de código são delimitados pela indentação, geralmente utilizando espaços em branco (tabs ou espaços). Ao contrário de outras linguagens que usam chaves `{}` para delimitar blocos, Python utiliza a indentação para indicar a estrutura do código. Isso não apenas torna o código mais legível, mas também garante que a hierarquia e a execução do programa sejam interpretadas corretamente pelo Python.

2. **Ordem de Execução (Leitura de cima para baixo)**: O Python executa o código de cima para baixo, linha por linha. Isso significa que a ordem em que as instruções são escritas é crucial para o fluxo de execução do programa. Variáveis devem ser definidas antes de serem utilizadas, e funções devem ser definidas antes de serem chamadas. Essa ordem é fundamental para garantir que o código funcione conforme esperado.

3. **Comentários**: Comentários em Python são marcados pelo caractere `#`. Eles são usados para explicar o código e fazer anotações que não são executadas pelo interpretador Python. Comentários são úteis para documentar partes complexas do código, explicar a lógica por trás de decisões específicas ou para tornar o código mais compreensível para outros desenvolvedores que possam revisá-lo no futuro.

4. **Instruções e Expressões**: Python suporta tanto instruções quanto expressões. Instruções são operações que realizam uma ação, como atribuições de variáveis, chamadas de função, estruturas de controle como `if`, `for`, `while`, entre outras. Expressões, por outro lado, produzem um valor, como operações aritméticas, chamadas de função que retornam um resultado, etc.

5. **Importação de Módulos**: Python permite importar funcionalidades de módulos externos para estender a funcionalidade padrão da linguagem. Isso é feito com declarações de importação, como `import nome_do_modulo` ou `from nome_do_modulo import funcao_especifica`. Importar módulos permite utilizar bibliotecas pré-existentes para realizar tarefas específicas, como manipulação de dados, operações matemáticas avançadas, entre outras.

6. **Organização e Estruturação**: Embora Python seja flexível quanto à estruturação do código, é uma boa prática organizar o código em funções e classes, quando necessário, para modularizar e reutilizar código. Funções permitem encapsular blocos de código que realizam tarefas específicas, enquanto classes fornecem um mecanismo para agrupar dados e funções relacionados em um único objeto.

Entender e aplicar corretamente esses conceitos de estrutura de código Python não apenas melhora a legibilidade e a manutenção do código, mas também facilita a colaboração entre desenvolvedores e aumenta a eficiência na criação de software robusto e funcional.
