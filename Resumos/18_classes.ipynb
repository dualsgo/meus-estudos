# **`Classes`**

À medida que construímos programas, muitas vezes precisaremos criar muitos objetos semelhantes, mas distintos. Por exemplo, vários cadastros de carros diferentes de uma concessionária.

Criar novas variáveis avulsas para cada um dos diferentes carros levaria muito tempo e poderia causar muita confusão.


````python
carro1_modelo = "Fusca"
carro1_ano = 1970

carro2_modelo = "Civic"
carro2_ano = 2022
````

````python
print("Modelo do Carro 1: " + carro1_modelo)
print("Ano do Carro 1: " + str(carro1_ano))
````

````python
print("Modelo do Carro 2: " + carro2_modelo)
print("Ano do Carro 2: " + str(carro2_ano))
````

Para nos ajudar a agrupar dados e funcionalidades, criamos uma classe. Uma classe é um 'molde' (ou modelo) que usamos para fabricar muitas coisas semelhantes, mas independentes.

A sintaxe é a seguinte:

```python
class NomeDaClasse:
    # código da classe
```

Usamos a indentação (espaços no início da linha) para indicar que o código abaixo da linha de definição pertence àquela classe.

O nome da classe deve começar com uma letra maiúscula. Por convenção (PEP 8), usamos CamelCase (ex: `ContaBancaria`) para nomes de classes em Python.

A classe `Carro` é um molde para criar carros. Ela não é um carro real estacionado na garagem, mas sim a planta de montagem para fabricar carros reais.


````python
class Carro:
    # O método __init__ é chamado quando criamos um novo objeto (fabricamos o carro). Ele nos permite definir as características iniciais.
    def __init__(self, modelo, ano):
        # O primeiro parâmetro de um método de classe é SEMPRE o 'self'. Ele se refere ao próprio objeto que está sendo criado.
        # Os outros parâmetros (modelo, ano) são as características que queremos salvar.

        self.modelo = modelo  # O atributo (característica) interno recebe o valor passado.
        self.ano = ano        
        
    # Podemos criar métodos (ações) para essa classe.
    def exibir_especificacoes(self):
        print("Modelo: " + self.modelo)
        print("Ano: " + str(self.ano))
        # Métodos nas classes são semelhantes a funções, mas são chamados usando um ponto (.).
````

Resumindo o que fizemos até agora: criamos a classe `Carro` como um molde. Esse molde define que todo carro terá atributos `modelo` e `ano`, e saberá usar a ação `exibir_especificacoes()`.

Ao usar esse molde, podemos criar quantos carros quisermos, sem criar variáveis genéricas soltas.


````python
carro_antigo = Carro("Fusca", 1970)  # Criamos um novo objeto da classe Carro e guardamos na variável carro_antigo.
# O 'self' é passado pelo Python automaticamente, só precisamos enviar 'Fusca' e 1970.

carro_novo = Carro("Civic", 2022)

print("Especificações do Carro Antigo:")
carro_antigo.exibir_especificacoes()  # Chamamos o método no objeto carro_antigo.
````

Criamos um método `exibir_especificacoes()` para facilitar a impressão. Mas também poderíamos ler as propriedades diretamente usando ponto, assim:


````python
print("Lendo diretamente do objeto:")
print("Qual o modelo do carro antigo? " + carro_antigo.modelo)
````

Vamos fazer um outro modelo passo a passo:

Desta vez, vamos criar uma classe `Pessoa` que tem valores padrões (sem usar o `__init__` no começo, apenas para ilustrar variáveis de classe):


````python
class Pessoa:
    print('Lendo o conteúdo dentro da classe...')
    
    # Criar variáveis numa classe funciona exatamente como criar variáveis normais.
    nacionalidade = 'Brasileiro(a)'
    hobby = 'Programar'

    # Adicionando uma funcionalidade (método).
    def apresentar(self):
        # Precisamos do 'self.' para acessar as variáveis da própria classe de dentro de um método.
        print('Olá, mundo!')
        print('Meu passatempo favorito é ' + self.hobby)
````

### Criando instâncias de classe

Quando usamos nosso molde (classe) para gerar um objeto real no programa, chamamos isso de **instanciar** e o resultado é uma **instância** (o objeto).


````python
class Mascote:
    cor = "caramelo"
    patas = 4
    vidas = 1
````

````python
rex = Mascote()
````

`Mascote` é a definição (o molde). `rex` é a instância (o objeto que criamos).

Para acessar uma propriedade do `rex`, usamos o ponto:


````python
print(rex.cor)  # caramelo
````

Podemos alterar o valor dessa propriedade para esse objeto específico:


````python
rex.cor = "preto"
rex.vidas = 7
print(rex.cor) # preto
````

Mas ficar mudando um por um depois de criar o mascote dá trabalho. É por isso que geralmente usamos construtores (o famoso `__init__`) para já nascerem diferentes!


````python
class Mascote:
    def __init__(self, cor, vidas):
        self.cor = cor
        self.vidas = vidas
````

Agora o molde EXIGE que digamos a cor e as vidas na hora de iniciar:


````python
rex = Mascote("preto", 7)
belinha = Mascote("caramelo", 1)
````

### Classes com métodos

E se quisermos que o mascote faça algum som?


````python
class Mascote:
    def __init__(self, nome, cor):
        self.nome = nome
        self.cor = cor

    def latir(self):
        print(self.nome + " diz: Au au!")
        
    def exibir_cor(self):
        print("A cor do " + self.nome + " é " + self.cor)
````

````python
rex = Mascote("Rex", "caramelo")
rex.latir()
rex.exibir_cor()
````

### Boas Práticas e Compreensão de Classes

Semelhante à nomeação de variáveis e funções, existem convenções (como a PEP 8) e boas práticas ao definir classes em Python:
1. Usar PascalCase/CamelCase (ex: `ContaBancaria`, `Pizza`, `Pessoa`).
2. Nomes descritivos de acordo com a entidade real ou abstrata.
3. Métodos podem usar todos os recursos do Python: laços `for`, `if/else`, etc.


````python
class Pizza:
    def __init__(self, sabor, ingredientes):
        self.sabor = sabor
        # ingredientes deve ser uma lista
        self.ingredientes = ingredientes

    def mostrar_receita(self):
        print("Para a pizza de " + self.sabor + " você precisa de:")
        for ingrediente in self.ingredientes:
            print("- " + ingrediente)
````

````python
pizza_marguerita = Pizza('Marguerita', ['massa', 'molho de tomate', 'mucarela', 'manjericão'])
pizza_marguerita.mostrar_receita()
````

### Métodos Especiais (Dunder Methods)

Existe outro método especial muito útil chamado `__str__`. Assim como o construtor `__init__`, ele é delimitado por dois sublinhados duplos (double underscores, ou *dunder* em inglês).

Por padrão, se você tentar imprimir um objeto como `print(pizza_marguerita)`, o Python imprime `<__main__.Pizza object at 0x...>`, o que não é muito legível.

O método `__str__` deixa você configurar que texto deve aparecer quando tentarmos converter o objeto para texto ou usa-lo no `print()`:


````python
class Mascote:
    def __init__(self, nome, cor):
        self.nome = nome
        self.cor = cor

    def __str__(self):
        # Este método deve retornar uma string!
        return f"Mascote(nome='{self.nome}', cor='{self.cor}')"

pet = Mascote("Floquinho", "verde")
print(pet)  # Saída bonita: Mascote(nome='Floquinho', cor='verde')
````

### Paradigmas: Programação Funcional vs Orientação a Objetos

Programação Funcional (FP) separa bem os dados das ações (funções):


````python
# Estilo Funcional
def calcular_distancia(velocidade_kmh, tempo_horas):
    return velocidade_kmh * tempo_horas

v = 100
t = 2
distancia = calcular_distancia(v, t)
print(distancia)
````

Na OOP (Orientação a Objetos), amarramos os dados e os métodos (ações) numa coisa só: o objeto.


````python
# Estilo OOP
class Viagem:
    def __init__(self, velocidade, tempo):
        self.velocidade = velocidade
        self.tempo = tempo
        
    def calcular_distancia(self):
        return self.velocidade * self.tempo

minha_viagem = Viagem(100, 2)
print(minha_viagem.calcular_distancia())
````

Em OOP, constantemente usamos os métodos para atualizar os dados do próprio objeto:


````python
class Cofrinho:
    def __init__(self):
        self.saldo = 0  
        
    def depositar(self, valor):
        self.saldo = self.saldo + valor

meu_cofre = Cofrinho()
meu_cofre.depositar(50)
meu_cofre.depositar(20)
print(meu_cofre.saldo)  # 70
````

### O que é o Encapsulamento?

Essência do encapsulamento é agrupar os dados pertinentes e suas respectivas responsabilidades no mesmo lugar. Mas não é só isso. Trata-se também de **ocultar e proteger** dados.

Muitas vezes, não queremos que outra parte do código modifique o interior do objeto de forma desajeitada. Em Python declaramos um atributo restrito (privado) com dois sublinhados `__` antes do nome.


````python
class ContaBancaria:
    def __init__(self, titular, saldo_inicial):
        self.titular = titular
        # Protegendo o saldo para que ninguém altere diretamente (ex: conta.__saldo = 999999)
        self.__saldo = saldo_inicial  
        
    def depositar(self, valor):
        if valor > 0:
            self.__saldo += valor
            
    def sacar(self, valor):
        if valor <= self.__saldo:
            self.__saldo -= valor
            
    def ver_saldo(self):
        return self.__saldo

conta_da_maria = ContaBancaria("Maria", 1000)
conta_da_maria.sacar(100)
print("Saldo real da conta:", conta_da_maria.ver_saldo())

# Se você rodar: print(conta_da_maria.__saldo)
# o Python apontará erro, indicando que este atributo não existe (proteção)!
````

### Aplicando Herança (Inheritance)

Se várias classes compartilham comportamentos parecidos, usar herança evita copiar e colar o mesmo código.


````python
class Pessoa:
    def __init__(self, nome):
        self.nome = nome
        
    def apresentar(self):
        print("Olá, eu sou " + self.nome)
````

Vamos criar um `Estudante` que vai "herdar" de `Pessoa`, afinal todo Estudante é uma Pessoa:


````python
class Estudante(Pessoa):
    # Ao colocar (Pessoa) na definição da classe, Estudante ganha acesso aos métodos da base
    def __init__(self, nome, curso):
        # super() nos dá acesso a "superclasse" que é a Pessoa. Chamamos o construtor dela!
        super().__init__(nome)
        self.curso = curso
        
    def exibir_curso(self):
        print("Eu estudo " + self.curso)

aluno = Estudante("Felipe", "Engenharia")
aluno.apresentar()  # Herdado de Pessoa!
aluno.exibir_curso() # Próprio de Estudante!
````

### Abstraindo objetos (Abstraction)

Abstração é focar no que é importante na visão de fora e esconder a complexidade de dentro.
Por exemplo, para operar uma máquina de café não precisamos conectar cada fio. Só apertamos o botão "fazer_cafe".


````python
class MaquinaDeCafe:
    def __init__(self):
        self.ligada = True
        
    def __aquecer_agua(self):   # Método privado! o usuário não chama isso.
        print("Aquecendo água a 90 graus...")
        
    def __moer_graos(self):     # Outro método privado.
        print("Moendo os grãos de café...")
        
    # Esse é o único método público que importa para quem vai beber o café:
    def preparar_cafe(self):
        if self.ligada:
            self.__aquecer_agua()
            self.__moer_graos()
            print("Café pronto! ☕")
        else:
            print("A máquina está desligada.")

maquina = MaquinaDeCafe()
maquina.preparar_cafe()  # Abstração na prática: chamamos 1 método simples e muito trabalho interno acontece!
````

### Polimorfismo (Sobrescrita de Métodos/Override)

As vezes um método existe na Classe Pai, mas precisamos que ele aja ligeiramente diferente dependendo da especialidade do Filho.
**Poli = Muitos | Morfos = Formas**. É a mesma ação, com formas diferentes de execução.


````python
class Animal:
    def emitir_som(self):
        print("Som misterioso...")
````

````python
class Cachorro(Animal):
    # O método 'emitir_som' já existia em Animal.
    # Ao recriar aqui, estamos sobrescrevendo o original!
    def emitir_som(self):
        print("Au Au!")
````

````python
class Gato(Animal):
    def emitir_som(self):
        print("Miau!")
````

````python
class Vaca(Animal):
    def emitir_som(self):
        print("Muuuu!")
````

````python
fazenda = [Cachorro(), Gato(), Vaca(), Animal()]

for bicho in fazenda:
    bicho.emitir_som()  # Mesma ação (polimorfismo), resultados diferentes!
````
